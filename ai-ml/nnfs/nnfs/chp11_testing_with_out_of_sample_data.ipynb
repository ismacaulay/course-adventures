### Testing with Out of Sample Data
**Overfitting** is effecively just memorizing the data without any understanding of it. Overfit models do very well prediciting data it has already seen but poorly for new data.



To identify if your model is overfitting, we need different data sets. **Training** data should be used to train your model, and **testing** (out of sample) data should be used to validate your models performance. One can take a random sampling of available data to train with, and make the rest testing data. This may cause information leakage which can be problematic for time series data. For time series data, a better approach is to take multiple slices of your data, entire blocks of time, and reserve them for testing.

We can recognize overfitting when testing data results start to diverge from training data result. When the loss differs by over 10% is a common sign of serious overfitting. Ideally, we want both data sets to have identical performance since even a small difference means that the modesl did not correctly predict some testing samples. Modest overfitting is not usually a problem, but something we want to minimize.

One general rule when selecting hyperparameters is to find the smallest possible model that still learns. The fewer neurons you have, the less chance that your network is memorizing the data. Other possible way to avoid overfitting is regulariztion, and the dropout layer.
