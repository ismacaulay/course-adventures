### Binary Logistic Regression

A model with an output layer that represents 2 classes, 0 or 1, is called **binary logistic regression**. A single neuron could represent cat vs dog, cat vs not cat, or any combination of 2 classes. Binary logistic regression is a regressor type of algorithm which will use a **sigmoid** activation function for the output rather than softmax. It will use **binary cross entropy** rather than bategorical cross entropy for calculating the loss.
