### L1 and L2 Regularization

Regularization methods reduce generalization error. L1 and L2 regularization are used to calculate a "penalty" which is added to the loss value. This penalty is used to penalize the model for large weights and biases. Large weights might indicate that a neuron is attempting to memorize the a data element, and instead it is generally believed that it is better to have many neurons contributing to a models output instead of a few. Usually, we only add regularization terms to the hidden layers (even if we are calling the method on the output layer since the values will be 0 and nothing will happen)
