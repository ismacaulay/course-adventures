## Optimizers

We can use the gradient to adjust the weights and biases to decrease the measure of loss. We subtracted a fraction of the gradient for each weight and biases parameter. This is a commonly used optimizer called **Stochastic Gradient Descent (SDG)**, and most optimizers are just varients of SGD.

In SDG, we choose a learning rate $\alpha$ and subtract $\alpha * parameter\_gradient$ from the parameter values. If the learning rate is 1, we are subtracting the actual gradient values from the parameters. How you choose the learning rate, and other hyper parameters, depends on the model, data, amount of data, parameter intialization method, etc. There is no single best way to set hyper parameters. The idea here is that we want to find the **global minimum** and avoid getting stuck in **local minimums**, but we dont know if the minimum we are in is global, or local



An **epoch** is a full pass through all of the trainging data during training. Usually, it requires multiple epochs to train a model, but the ideal scenario would be to have a perfect model with ideal weights and biases after a single epoch. To add multiple epochs, we just loop and perform the forward, backward, and optimization passes.

### Stochastic Gradient Descent with Learning Rate Decay

The idea of a **learning rate decay** is to start with a large learning rate and then decrease it during training. There are a few ways of doing this:
1. decrease the learning rate in response to the loss across epochs
2. add a decay rate with steadily decays the learning rate per batch or epoch

Decaying per step is also referred to as **1/t decaying** or **exponential decaying**, which basically means we will update the learning rate each step by the reciprocal of the step count fraction. This fraction is another hyperparameter that will be added to the optimizer. This works by taking the step and the decaying ratio and multiplying them, which means the furthur into training, the bigger the step, which in terms means a bigger multiplication. We then use the reciprical of that factor plus 1 (to ensure it never raises the learning rate).



Stochastic gradient descent with learning rate decay can do fairly well, but it is still fairly basic and we can do better!

### Stochastic Gradient Descent with Momentum

**Momentum** creates a rolling average of gradients over some number of updates and uses this average with the unique gradient at each step. It uses the previous updates direction to influence the next updates direction to help minimize the chance of bouncing around and getting stuck in a local minimum.

We use momentum by setting another hyper parameter between 0 and 1 that represents the fraction of the previous parameter update to retain, and subtracting the update from it.



If we set the momentum parameter to high, the model might stop learning since the direction of the updates wont be able to follow the global gradient descent
