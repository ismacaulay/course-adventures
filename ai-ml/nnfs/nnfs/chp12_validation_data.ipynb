### Validation Data

Hyperparameter tuning should not be done using the test data. Instead, we should us a **validation** data set. The test dataset should only be used as unseen data to test model performance. It should not influence the model in anyway. If we have alot of training data, we can afford to use some for a validation data set.

If we are short on data, we have have 2 options:

1. Temporally split the training data into a smaller training dataset and validation dataset. Afterwars, with the final hyperparameters set, train the model on all of the training data.

2. Use a process call cross validation by splitting the data set into multiple parts. Then train the model on the first n-1 parts and validate on the last one. Then swap the parts around until you have validated on each part. This way we do not lose any training data. This is also called k-fold cross-validation
